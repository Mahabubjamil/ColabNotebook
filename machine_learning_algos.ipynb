```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("credit card.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["AGE"].fillna(df["AGE"].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
X = df.drop(["default.payment.next.month"],axis=1)
```

```python
X.head()
```

```python
Y = df["default.payment.next.month"]
Y.head()
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_features = SelectKBest(score_func = f_classif, k ="all")
```

```python
fit_features.fit(X,Y)
```

```python
df1 = pd.DataFrame(fit_features.scores_,index = X.columns, columns = ["Score"] )
```

```python
df1.nlargest(24,["Score"])
```

```python
from sklearn.svm import SVC
sv = SVC(gamma='auto')
sv.fit(X_train,Y_train)
```

```python
pred1 = sv.predict(X_test)
```

```python
sv.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(Y_test,pred1)
```

```python
print(cr)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
```

```python
bnb.fit(X_train,Y_train)
```

```python
pred2 = bnb.predict(X_test)
```

```python
bnb.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(Y_test,pred2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train,Y_train)
pred3 = dt.predict(X_test)
```

```python
dt.score(X_test,Y_test)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(Y_test,pred3)
print(cr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rnr = RandomForestClassifier(n_estimators= 50)
```

```python
rnr.fit(X_train,Y_train)
pred4 = rnr.predict(X_test)
```

```python
rnr.score(X_test,Y_test)*100
```

```python
from sklearn.metrics import classification_report
cr = classification_report(Y_test,pred4)
print(cr)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.metrics import classification_report
```

```python
conf = confusion_matrix(Y_test,pred4)
print(conf)
```

```python
import seaborn as sns
%matplotlib inline

class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(conf), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix of Random Forest', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```
