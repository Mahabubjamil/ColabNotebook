***Time Resampling***

```python
df_tesla.head()
```

```python
#A meansyear and frequency
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()['Open'].plot(figsize=(12,4))
```

```python
#QS means qutarly
df_tesla.resample(rule='QS').min()
```

```python
#ploting
df_tesla["Open"].resample(rule='QS').min().plot(kind="bar")
```

```python
df_tesla["Open"].resample(rule='QS').mean().plot(kind='bar',figsize=(15,6))
```

***Rolling***

```python
df_tesla['High'].rolling(10).mean().head(20)
```

```python
df_tesla['open:10 days rolling'] = df_tesla['Open'].rolling(10).mean()
```

```python
df_tesla.head(15)
```
