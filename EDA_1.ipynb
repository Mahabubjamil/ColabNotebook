```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on="Country Code", how = "left")
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_name = final_df.Country.value_counts().index
country_value = final_df.Country.value_counts().values
plt.pie(country_value[0:5],labels=country_name[:5],autopct='%1.2f%%')
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data=ratings,palette=["blue","red","orange","yellow","green"])
```

observation: 1. not rated count is very high
2. max no of reating ae between 2.8 to 3.9.

```python
#count plot
sns.countplot(x="Rating color", data=ratings,palette=["blue","red","orange","yellow","green"])
```
