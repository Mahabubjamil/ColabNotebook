```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
housing
```

```python
housing
```

```python
datasets = pd.DataFrame(housing.data)
```

```python
datasets.columns = housing.feature_names
datasets.head()
```

```python
datasets['Price'] = housing.target
```

```python
datasets.head(10)
```

```python
#dividing the datasets into dependent and independent features
X = datasets.iloc[:,:-1]
X.head()
```

```python
Y = datasets.iloc[:,-1]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=2)
```

```python
#linear Regrassion
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lr = LinearRegression()
lr.fit(X_train,Y_train)
mse = cross_val_score(lr,X_train,Y_train,scoring='neg_mean_squared_error',cv = 5)
print(mse)
```

```python
mean_mse = np.mean(mse)
mean_mse
```

```python
y2_pred = lr.predict(X_test)
from sklearn.metrics import r2_score
r2_score2 = r2_score(y2_pred,Y_test)
print(r2_score2)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()
params = {'alpha':[1e-15,1e-10,1e-8,1e-5,1e-3,1e-2,1,2,5,8,10,20]}
ridge_regressor = GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,Y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
#R-squared
y1_pred = ridge_regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score2 = r2_score(y1_pred,Y_test)
print(r2_score2)
```

```python
#Lasso Regression
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
params = {'alpha':[1e-15,1e-10,1e-8,1e-5,1e-3,1e-2,1,2,5,8,10,20]}
lasso_regressor = GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(X_train,Y_train)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
y_pred = lasso_regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score1 = r2_score(y_pred,Y_test)
print(r2_score1)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
X = pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
X.head()
```

```python
y = pd.DataFrame(df['target'],columns=['Target'])
y.head()
```

```python
y["Target"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
params=[{'C': [1,5,10]},{'max_iter':[100,150]}]
```

```python
model1 = LogisticRegression(C=100,max_iter=100)
```

```python
model = GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
